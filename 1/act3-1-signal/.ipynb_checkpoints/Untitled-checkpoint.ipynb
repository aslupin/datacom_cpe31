```python
import urllib
from sigproc import Signal,start_notebook
start_notebook()

urllib.request.urlretrieve("https://www.cpe.ku.ac.th/~cpj/204325/q3-1/q3-1-58.wav","q3-1-58.wav")
s = Signal(wav_file="q3-1-58.wav")
s.plot_time(fig_options=dict(x_range=(0,10)))
s.plot_freq()
s.plot_phase()
```

```python
from numpy import cos,pi

def f(t):
    return (0.7 * cos(2*pi*300*t - 3.14))



s2 = Signal(func=f)
s2.plot_time(fig_options=dict(x_range=(0,10)))
```

```python
from numpy import cos,pi,radians

def f(t):
    return (
         (0.2 * cos(2*pi*600*t + radians(180))) + 
         (0.3 * cos(2*pi*800*t + radians(90))) +
         (0.1 * cos(2*pi*900*t + radians(270)))
    )


s2 = Signal(func=f)
s2.plot_time(fig_options=dict(x_range=(0,100)))
```

```python
import numpy
def getdB(x):
    return 10 * numpy.log10(x)
def getCapacity(bw, snr):
    return bw * numpy.log2(1 + snr)
    
print(getdB(100/85))
print(getdB(3333.33))
print(getCapacity(4000, 3333.33) / 1000)
print(getCapacity(4000, 3333.33) / 2)
```

```python
import numpy as np
import matplotlib.pyplot
LIGHT_SPEED = 3e8
DOG_SPEED = 4*1000/3600   #  m/s
size = np.arange(1,10000,1)
def method1_time(size):  # 
    #fiberTransfering = 113 / LIGHT_SPEED
    size = size * 8
    transfering = size / 88 
    return transfering

def method2_time(size):  # 
    runing =   113  / DOG_SPEED
    pollingReadUSB = size / 245
    pollingWriteUSB = size / 190
    return pollingReadUSB + runing + pollingWriteUSB

print(method1_time(188 * 8))
print(method2_time(188))

pyplot.plot(size,method1_time,'k')
```
