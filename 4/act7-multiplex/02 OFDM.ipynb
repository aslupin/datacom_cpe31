# OFDM Example
โน้ตบุ๊คนี้เตรียมโค้ดอย่างง่ายเพื่อเป็นตัวอย่างให้เห็นถึงสัญญาณที่มีความถี่ orthogonal กัน (orthogonal frequency) ซึ่งเป็นกลไกสำคัญที่นำมาใช้ในกระบวนการ Orthognal Frequency Multiplexing (OFDM)

```python
from numpy import sin,cos,pi
from sigproc import Signal,start_notebook,DEFAULT_PLOT_SETTINGS
from sigproc.analog import Qam

bkp = start_notebook()
DEFAULT_PLOT_SETTINGS['time_freq_fig_options']['height'] = 200
DEFAULT_PLOT_SETTINGS['freq_fig_options']['x_range'] = (0,1000)
```

## สร้างมอดูเลเตอร์และมอดูเลทข้อมูล 2 ชุด

```python
CONSTELLATION = {
    '0' : (0.3,0), 
    '1' : (0.3,180) 
}
BAUD = 10
FC1 = 500
FC2 = 600
data1 = '0100000000000000'
data2 = '0100000000000000'

# print("Data stream #1")
qam1 = Qam(CONSTELLATION,bitrate=BAUD*1,fc=FC1)
sig1 = qam1.generate_signal(data1)
sig1.plot_time_and_freq()

# print("Data stream #2")
qam2 = Qam(CONSTELLATION,bitrate=BAUD*1,fc=FC2)
sig2 = qam2.generate_signal(data2)
sig2.plot_time_and_freq()
```
