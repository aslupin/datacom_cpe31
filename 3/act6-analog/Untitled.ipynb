```python
import urllib
from sigproc import Signal,start_notebook
from sigproc.analog import Qam
bkp = start_notebook()

urllib.request.urlretrieve("https://www.cpe.ku.ac.th/~cpj/204325/q6-3/q6-3-37.wav","q6-3-37.wav")
s = Signal(wav_file="q6-3-37.wav")
s.plot_time()
s.plot_freq()


carrier_i = Qam.generate_frequency(580,s.duration,phase=0)
(carrier_i * s * 2).filter(580).plot_time()

carrier_q = Qam.generate_frequency(580,s.duration,phase=90)
(carrier_q * s * 2).filter(580).plot_time()
```
